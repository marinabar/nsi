<h1>1er partie : affichage des points (1 h)</h1>
<p>Nous allons écrire une fonction <b>charger_data</b> qui 
   va importer les données issues du fichier csv et retourner une 
   liste de tuples de la forme (x,y,lab). <br>
   Les paramètres LongueurPetale , LargeurPetale et Variete sont 
   respectivement attribués aux variables x, y et lab.<br></p>  
<ol>
   <li>Réaliser la fonction "charger_data". Exécuter la 
      procédure pour la tester.
   </li>
</ol>

```python
def charger_data():
    data=[]
    file=open("iris.csv","r")
    file.readline() #extraire la première ligne
    iris=file.read().strip().split('\n')    #nettoyer et séparer les coordonnées des points
    
    for i in range (len(iris)):
        data.append(list(map(int, iris[i].split(','))))
    
    file.close()
    return data
```

<h1>2ième partie : k plus proches voisins (1 h)</h1>
<p>Nous allons implanter le programme qui permet de connaître les k 
   plus proches voisins d'un point donné puis d'en déduire la variété 
   de l'iris.</p>
<ol start="4">
   <li>Écrire la fonction <b>distance</b> qui prend comme arguments 
      deux tuples de coordonnées et qui renvoie la distance.</li>
   <li>Écrire la fonction <b>listdist</b> qui prend comme 
      arguments la liste de tuples de coordonnées et un tuple et 
      qui renvoie une liste de distances catégorisée (liste de [d, lab]).</li>
</ol>

```python
def distance(a, b):
        
    d =0
    for i in range (len(a)):
        d+=(a[i]-b[i])**2
    return sqrt(d)
    
```

```python
def listdist(data, o):
       
    listd = []
    for i in range (len(data)):
        listd.append((distance(data[i][:2], o), data[i][2]))
    return listd
    
```

```python
'''Code pour tester les deux fonctions précédentes'''
point=(48,17)
ld=listdist(table,point)
print(ld)
```

<ol start="6">
   <li>Écrire la fonction <b>kppv</b> qui prend comme arguments 
      la liste de distances catégorisée, un tuple de coordonnées et un entier k 
      et qui renvoie une liste des k plus proches voisins.</li>
</ol>

```python
def kppv(ld, k, o):
    d = sorted(ld, key=lambda x:x[0])# trier liste en fonction du 1er arg dans chaque tuple de la liste
    return d[:k]
```

```python
'''Code pour tester la fonction'''
point=(48,17)
ld=listdist(table,point)
k=3
lk=kppv(ld,k,point)
print(lk)
```

<ol start="7">
    <li>Écrire la fonction <b>prediction</b> qui prend comme arguments 
      la liste de distances catégorisée et qui renvoie la variété 
      d'iris majoritaire.</li>
</ol>

```python
def prediction(lk):
    final = [k[1] for k in lk]
    return (max(set(final), key=final.count))
```

```python
'''Programme principal'''
plt.clf()
table = charger_data()
afficher(table)
point=(48,17)
ld=listdist(table,point)
k=3
lk=kppv(ld,k,point)
ck=prediction(lk)
afficher_point(point,ck)
plt.text(50, 5,  f"catégorie : {labels[ck]}",fontsize=8, bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1'))
plt.show()
```
